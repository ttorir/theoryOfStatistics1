```python
import pandas as pd
import matplotlib.pyplot as plt
import graphviz
from semopy import Model, gather_statistics
from matplotlib.colors import ListedColormap
import seaborn as sns
import numpy as np

plt.style.use('fivethirtyeight')

pd.set_option('display.max_columns', None)
```

```python
cases_dict = {'Case I': 'case_I.pkl',  
                'Case II': 'case_II.pkl'}
case_disp = 'Case II'

table = pd.read_pickle(cases_dict.get(case_disp))
table.columns = [' '.join(x.split('_')).title() for x in table.columns]
table.index = ['AIC','AICc','BIC','GFI',f'$R^2$',f'Adjusted $R^2$']
```

```python
plt.figure(figsize=(8,5))
ax=sns.heatmap(table.div(table.sum(axis=1), axis=0),cmap="YlGn", annot=True, fmt='.1%',cbar=False)
ax.figure.axes[-1].yaxis.label.set_size(14)
plt.title(f'{case_disp}: Model Selection by Method', pad=30,fontweight="bold")
plt.ylabel('Selection Method')
plt.tick_params(axis='both', which='major', labelsize=12, labelbottom = False, bottom=False, top = False, labeltop=True)
# plt.tight_layout()
ax.set_facecolor('white')
plt.show()
```
